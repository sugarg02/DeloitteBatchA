Given a list of numbers, write a Python program to count positive and negative numbers in a List.

Input: list1 = [2, -7, 5, -64, -14]
Output: pos = 2, neg = 3

```python
lst = []
n = int(input("Enter the number of elements:"))
for i in range(n):
  ele=int(input())
  lst.append(ele)
print(lst)
```

```python
 list1 = [2, -7, 5, -64, -14]
 pos =0
 neg=0
 for num in list1:
   if num >= 0:
     pos=pos+1
   else:
      neg=neg+1
print("positive nunbers count is:", pos)
print("negative nunbers count is:", neg)
```
